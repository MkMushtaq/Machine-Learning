```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('canada_per_capita_income.csv')
```

```
df.head()
```

```
import matplotlib.pyplot as plt
plt.scatter(df['year'], df['per capita income (US$)'], color = 'r') 
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(df['year'], df['per capita income (US$)'], test_size = 0.2)
```

```
from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression()
model.fit(np.array(x_train).reshape(37,1), np.array(y_train).reshape(37,1))
```

```
y_pred = model.predict(np.array(x_test).reshape(10,1))
y_pred
```

```
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
