```
iterations = 1000000
lr = 0.0002
m = 0
b = 0
for i in range(iterations):
  dm = -(2/n)*(sum(x*(y - (m*x + b))))
  db = -(2/n)*(sum((y - (m*x + b))))
  m = m - lr*dm
  b = b - lr*db
  cost = (1/n)*sum(np.square(y  - (m*x + b)))
  print('iteration:',i, 'm:', m,'b:', b, 'cost:', cost)
```
