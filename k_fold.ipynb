```
import numpy as np, pandas as pd
```

```
from sklearn.datasets import load_iris
df = load_iris()
```

```
x = df['data']
y = df['target']
```

```
from sklearn.model_selection import train_test_split, cross_val_score
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```
def get_score(model, x_train, x_test, y_train, y_test):
  score = cross_val_score(model, x_train, y_train, cv = 4)
  return score
```

```
from sklearn.linear_model import  LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```
get_score(LogisticRegression(solver='liblinear'),x_train, x_test, y_train, y_test)
```

```
get_score(SVC(),x_train, x_test, y_train, y_test)
```

```
get_score(RandomForestClassifier(n_estimators=40),x_train, x_test, y_train, y_test)
```
