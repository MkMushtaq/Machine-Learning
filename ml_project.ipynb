```
def is_float(x):
    try:
      float(x)
    except:
      return False 
    return True
```

```
def get_float(x):
  t = x.split('-')
  if len(t)==2:
    return (float(t[0]) + float(t[1]))/2
  try:
    return float(x)
  except: 
    return None
```

```
def get_scores(x_train, y_train):
    algos = {
        # 'Linear Regression': {
        #     'model': LinearRegression(),
        #     'params': {
        #         'normalize': False
        #     }
        # },
        'Random Forest': {
            'model': RandomForestClassifier(),
            'params': {
              'n_estimators': [1, 5, 10, 20, 40]
            }
        }

    }

    scores = []

    for i,config in algos.items():
        gs =  GridSearchCV(config, config['params'])
        gs.fit(x_train, y_train)
        scores.append({
            'model': i,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])
```
