```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv('carprices.csv')
df.head()
```

```
OHE_values = pd.get_dummies(df['Car Model'])
OHE_values.drop(columns = ['Mercedez Benz C class'], inplace = True)
```

```
df_new = pd.concat([df, OHE_values], axis = 1).drop(columns = 'Car Model')
df_new
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
df_new.head()
```

```
target_column = ['Sell Price($)']
columns = df_new.columns
non_target_columns = list( set(columns) - set(target_column) )
non_target_columns
```

```
model.fit(df_new[non_target_columns], df_new[target_column])
```

```
model.predict([[0,1,7,86000]])
```

```
model.score(df_new[non_target_columns], df_new[target_column])
```
